A continuacin separamos todos los gneros de pelculas del conjunto de datos y use la estructura de datos "establecida" para no duplicar gneros en el mismo. 

Si en la columna "|" encuentra, separar los gneros de las pelculas y eliminar la duplicacin en el mismo mediante el uso de la estructura de datos "establecida".

```python
genre_list=" "
for index,row in movies.iterrows():
    genre_list+=row.genres+"|"
genre_list_split=genre_list.split("|")
new_list=list(set(genre_list_split))
new_list.remove('')
new_list
```

```python
m=movies.copy()
```

A continuacin lo que quiero hacer es pasar a un valor numrico cada pelcula a que genero pertenece pero en dimensiones separadas

```python
for genre in new_list:
    m[genre]=m.apply(lambda _:int(genre in _.genres),axis=1)
m.head()
```

De a cuerdo a la consigna me tengo que quedar con las pelculas que obtuvieron un rating >= 4

```python
highratings=ratings[ratings.rating>=4]
highratings
```

Anteriormente obtuve los cdigos de las pelculas con rating >=4 pero tamb necesito el nombre de las pelculas para poder recomendar y que el usuario entienda

```python
itemsets=[[idx2title[mov] for mov in highratings[highratings.userId==user].movieId]
         for user in highratings.userId]
itemsets
```

Una vez que tengo los datasets armados puedo invocar al algoritmo Apriori para el sistema de recomendacin.

Lo que quiero hacer a continuacin es lo siguiente:
  - Por cada pelicula (Item) quiero poner si el usuario 1....x consumi determenida pelcula...si la vio va a tener un True si no la vi va a tener un False.

Me transforma los items en este caso las pelculas y en las filas los usuarios
    

Si obtienen error con la librera Module not found mlxtend - la instalamos con pip install mlxtend

```python
from mlxtend.preprocessing import TransactionEncoder #pip install mlxtend
te=TransactionEncoder()
tr_ary=te.fit(itemsets).transform(itemsets)
DF=pd.DataFrame(tr_ary,columns=te.columns_)
DF.head()
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

Con el parmetro use_colnames = True le estoy indicando que solo tome las columnas que tengas = True que significa lo que vio el usuario

```python
f=apriori(DF, min_support=0.2, use_colnames=True, max_len=2)
#Esto es una variante a lo que vimos la clase anterior
rules=association_rules(f,metric='lift',min_threshold=2)
rules.head()
```

Sobre lo anterior, se ha recomendado que la pelcula use el usuario donde 'antecedentes' es la columna de la pelcula y 'consecuencias' es la columna de la pelcula recomendada. 

En la primera fila, si ve la pelcula '(2001:A Space Odyssey(1968))', se le recomendar '(Blade Runner(1982))' para ver ms y cuya confianza de la vista es del 72 % aproximadamente. 

La columna lift se usa para decirnos que la probabilidad de ver ambas pelculas juntas es 2,144151 veces mayor que la probabilidad de ver solo una pelcula. El soporte es la popularidad predeterminada de un artculo. 

En trminos matemticos, el soporte de un artculo no es ms que la proporcin de transacciones que involucran un artculo al nmero total de transacciones. support(movie_1)=(Todas las transacciones que involucran movie_1)/(total de transacciones) 

Conviction compara la probabilidad de que X aparezca sin Y si fueran dependientes con la frecuencia real de aparicin de X sin Y. 

El umbral en el algoritmo Apriori identifica los conjuntos de elementos que son subconjuntos de al menos ya que cada transaccin se ve como un conjunto de elementos.

El leverage calcula la diferencia entre la frecuencia observada de A y C que aparecen juntos y la frecuencia que se esperara si A y C fueran independientes. Un valor de leverage de 0 indica independencia.
