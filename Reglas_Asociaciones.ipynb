```python
df1 = pd.read_csv('ratings.csv')
df3 = df1[df1['rating'] >= 4]
```

```python
df2 = pd.read_csv('movies.csv')
print(df2.columns.values)
df4 = df2[['movieId', 'title']]
```

```python
merged_inner = pd.merge(left=df3, right=df4, left_on='movieId', right_on='movieId')
user2 = set(list(merged_inner.userId))
titles = (merged_inner.title.tolist())
user = set(merged_inner.userId.tolist())
```

```python
# Para crear los títulos sin duplicados
titles2 = []
for title in titles:
    if title not in titles2: 
        titles2.append(title)
print(titles2)
```

```python
# Para crear lista de usuarios
user2 = []
for i in user:
    if user not in user2:
        user2.append(i)
print(user2)
```

```python
#convertir lista a DataFrame
df = pd.DataFrame (columns= [titles2], index= [user2])
df
```
